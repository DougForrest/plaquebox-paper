```python
results = utils.load_json(os.path.join(cfg.output_path, 'rn50-stage2-v1_2019_12_22.json'))
```

```python
def results_to_df(results):

    df = pd.DataFrame.from_dict(results['results'])
    df.index.name = 'metric'
    df['experiment'] = results['meta']['experiment']
    df['model'] = results['meta']['model']
    df['date'] = datetime.datetime.strptime(results['meta']['date'], '%Y_%m_%d').date()
    df.reset_index(inplace=True)
    
    return (df)

    
```

```python
a = results_to_df(results)
```
